```python
from bs4 import BeautifulSoup
from requests import get
from time import sleep
from random import randint
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
headers = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})
```

```python
sapo = "https://casa.sapo.pt/Venda/Apartamentos/?sa=11&or=10"
response = get(sapo, headers=headers)
```

```python
print(response)
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
house_containers = html_soup.find_all('div', class_="searchResultProperty")
```

```python
#using itertools to retrieve the digits and turn it into a float
var_1 = int(''.join(itertools.takewhile(str.isdigit, var_1)))
print(var_1, type(var_1))
```

```python
# setting up the lists that will form our dataframe with all the results
titles = []
created = []
prices = []
areas = []
zone = []
municipality = []
city = []
condition = []
descriptions = []
urls = []
thumbnails = []
```

```python
%%time

n_pages = 0

for page in range(0,900):
    n_pages += 1
    sapo_url = 'https://casa.sapo.pt/Venda/Apartamentos/?sa=11&lp=10000&or=10'+'&pn='+str(page)
    r = get(sapo_url, headers=headers)
    page_html = BeautifulSoup(r.text, 'html.parser')
    house_containers = page_html.find_all('div', class_="searchResultProperty")
    if house_containers != []:
        n_container = 0
        for container in house_containers:
            n_container+=1
            #print(n_pages, " ", n_container)
            
            # Price            
            price = container.find_all('span')[2].text
            if price == 'Contacte Anunciante':
                price = container.find_all('span')[3].text
                if price.find('/') != -1:
                    price = price[0:price.find('/')-1]
            if price.find('/') != -1:
                price = price[0:price.find('/')-1]
            
            price_ = [int(price[s]) for s in range(0,len(price)) if price[s].isdigit()]
            price = ''
            for x in price_:
                price = price+str(x)
            if(price!=''):
                prices.append(int(price))
            else:
                prices.append(price)
                
            # Zone
            location = container.find_all('p', class_="searchPropertyLocation")[0].text
            location_zone = location[7:location.find(',')]
            zone.append(location_zone)
            # Municiplaities
            if(location.count(',')>1):
                location_municiplaity = location[location.rfind(',',0,location.rfind(',')-1)+1:location.rfind(',')]
            municipality.append(location_municiplaity)
            # City
            location_city = location[location.rfind(',')+1:]
            city.append(location_city)
            
            # Title
            name = container.find_all('span')[0].text
            titles.append(name)
            
            # Status
            status = container.find_all('p')[5].text
            condition.append(status)
            
            # Area
            m2 = container.find_all('p')[9].text
            if m2 != '-':
                m2 = m2.replace('\xa0','')
                m2 = float("".join(itertools.takewhile(str.isdigit, m2)))
                areas.append(m2)
                
            else:
                m2 = container.find_all('p')[7].text
                if m2 != '-':
                    m2 = m2.replace('\xa0','')
                    m2 = float("".join(itertools.takewhile(str.isdigit, m2)))
                    areas.append(m2)
                else:
                    areas.append(m2)
            
            # Creation date
            try:
                date = pd.to_datetime(container.find_all('div', class_="searchPropertyDate")[0].text[21:31])
            except:
                date = None
            created.append(date)
            
            # Description
            desc = container.find_all('p', class_="searchPropertyDescription")[0].text[7:-6]
            descriptions.append(desc)

            # url
            link = 'https://casa.sapo.pt/' + container.find_all('a')[0].get('href')[1:-6]
            urls.append(link)

            # image
            img = str(container.find_all('img')[0])
            img = img[img.find('data-original_2x=')+18:img.find('id=')-2]
            thumbnails.append(img)
    else:
        break
    
    sleep(randint(1,2))
    
print('You scraped {} pages containing {} properties.'.format(n_pages, len(titles)))
```

```python
cols = ['Title', 'Zone', 'Municipality', 'City', 'Price', 'Size (m²)', 'Status', 'Description', 'Date', 'URL', 'Image']

lisboa = pd.DataFrame({'Title': titles,
                           'Price': prices,
                           'Size (m²)': areas,
                           'Zone': zone,
                           'Municipality': municipality,
                           'City': city,
                           'Date': created,
                           'Status': condition,
                           'Description': descriptions,
                           'URL': urls,
                           'Image': thumbnails})[cols]

lisboa.to_excel('lisboa_raw.xls', index=False)
```

```python
lisboa_clean['Size (m²)'] = pd.to_numeric(lisboa_clean['Size (m²)'], errors='coerce')
lisboa_clean.dropna(subset=['Size (m²)'], inplace=True)
```

```python
lisboa_clean["Zone"] = lisboa_clean["Zone"].str.strip()
lisboa_clean["Municipality"] = lisboa_clean["Municipality"].str.strip()
lisboa_clean["City"] = lisboa_clean["City"].str.strip()
```

```python
lisboa.drop(['URL', 'Image'], axis=1, inplace=True)
```

```python
rooms = []
for h in lisboa_clean.Title:
    if(type(h)==str or not math.isnan(h)):   
        n = [(s) for s in re.findall(r'T\d{1,2}',h)]
        rooms.append(n)
roomsT = pd.DataFrame(rooms)[0]

roomsT.unique()
```

```python
lisboa_clean['Rooms'] = roomsT
print(len(lisboa_clean['Rooms']))
lisboa_clean.dropna(subset=['Rooms'], inplace=True)
lisboa_clean.reset_index(inplace=True, drop=True)
print(len(lisboa_clean['Rooms']))
```

```python
r=[]
for x in lisboa_clean['Rooms']:
    r.append(int(x[1:]))

lisboa_clean['Rooms'] = r
```

```python
lisboa_clean['Price/m²'] = lisboa_clean['Price'] / lisboa_clean['Size (m²)']
```
